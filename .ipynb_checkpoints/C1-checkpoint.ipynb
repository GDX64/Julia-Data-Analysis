# TFP time series analysis

We are going to analyse the time series in the TFP dataset

```julia
using CSV
using DataFrames
using Plots
using LinearAlgebra
using StatsBase
using Statistics
```

```julia
tfp=CSV.read("TFP.csv");
```

```julia
tfp[1:5,:]
```

```julia
gp=groupby(tfp, :isocode);
```

```julia
@show typeof(gp[1])
for g in gp 
    print(describe(g))
end
```

```julia
combine(gp, :rtfpna=>mean=>"mean", 3=>std=>:std, 3=> begin x->max(x...) end => "max",
    3=> begin x->min(x...) end => "min")
```

### Correlation between countries 

There are no strong correlation between countries

```julia
A=[gp[1][:,end] gp[2][:,end] gp[3][:,end]]

cor(A)
```

As the time series is relatively short, this kind of numbers don't say much. Making a lineplots would be more insightfull

## Ploting

```julia
p=plot()
for g in gp
    plot!(g[:,end-1],g[:,end], label=g[1,1], ylabel="TFP", title="Countries TFP time series", xlabel="Years")
end
p
```

```julia
p1=plot(); p2=plot(); p3=plot()

for g in gp
    v=collect(g[!,end])
    plot!(p1, autocor(v.-mean(v)), label=g[1,1]) 
    v=diff(v)
    plot!(p2, autocor(v.-mean(v)), label=nothing)
    v=diff(v)
    plot!(p3, autocor(v.-mean(v)), label=nothing) 
end
title!(p1, "Correlation of V"); title!(p2,"Correlation of V'"); title!(p3,"Correlation of V''")
l = @layout [a ; b c]
plot(p1,p2,p3, layout=l)
```

By the shape of the plots, we can say that Cannada and Mexico are not proper for using good old linear regression. They don't show any seasoning pattern. We could roll a mean and std to the graph but we can see with naked eye that the series does not fit a starionary model. We can use an ARIMA model to predict the next samples, but it will need to have at least d=2

```julia
model=ar.ARIMA(gp[1][1:50,3], order=(1,2,1))
model_fit=model.fit(disp=0)
const N_fsamples=10
```

```julia
using Measurements
```

## Ploting all three

```julia
#This is an auxiliar forecasting function
function trainAndFC(df)
    model=ar.ARIMA(df,order=(2,2,2))
    model_fit=model.fit(disp=0)
    return model_fit.forecast(N_fsamples)
end
```

```julia
myPlots=[]
colors=["blue", "green", "black"]

for (i,g) in enumerate(gp)
    forecast, stderr, conf =trainAndFC(g[1:end,3])
    N=length(g[1:end,3])
    
    p=plot()
    
    plot!(g[1:end,3], label=g[1,1], color=colors[i])
    
    forePlot=forecast .± stderr
    
    plot!([(1:N_fsamples+1).+(N-1)], [g[end, 3]; forePlot], label=nothing, color=colors[i], ylabel="TFP", xlabel="Years")
    
    push!(myPlots, p)
end
title!(myPlots[1], "Forecasting")
plot(myPlots..., layout=(3,1), size=(800, 800))
```

## Conclusion

The model can do some prediction, though it is hard to believe in the 10 years forecast, the trend seems to play a big role in the final results.
